# Data Preparation

```python
init_data = pd.read_csv("london_weather.csv", index_col="date")
```

```python
init_data.head()
```

```python
# Convert index column to a DateTime format
init_data.index = pd.to_datetime(init_data.index, format="%Y%m%d")
init_data
```

```python
nan_percentage = init_data.apply(pd.isnull).mean()
nan_percentage
```

```python
init_data.index.year.value_counts().sort_index()
```

```python
init_data["target"] = init_data["mean_temp"].shift(-1)
init_data
```

```python
weather_data = init_data.ffill()
weather_data
```
